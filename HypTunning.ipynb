```python
from sklearn.datasets import load_iris
from sklearn import svm
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
iris=load_iris()
dir(iris)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

```python
iris.target_names
```

```python
df['target']=iris.target
df['target']=df['target'].apply(lambda x:iris.target_names[x])
df
```

```python
X_train, X_test, y_train, y_test=train_test_split(df.drop(columns=['target']),df.target,test_size=0.3)
X_train
```

```python
model=svm.SVC(kernel='rbf',C=30,gamma='auto')
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np
kernels=['rbf','linear']
c=[1,10,20]
avg_scores={}
for ker in kernels:
    for cval in c:
        cv_scores=cross_val_score(svm.SVC(kernel=ker,C=cval,gamma='auto'),df.drop(columns=['target']),df.target,cv=5)
        avg_scores[ker+'_'+str(cval)]=np.average(cv_scores)
avg_scores
```

```python
from sklearn.model_selection import GridSearchCV
crvl=GridSearchCV(svm.SVC(gamma='auto'),{'kernel':['rbf','linear'],'C':[1,10,20]},cv=5,return_train_score=False)
crvl.fit(df.drop(columns=['target']),df.target)
crvl.cv_results_
```

```python
results=pd.DataFrame(crvl.cv_results_)
results
```

```python
results[['param_C','param_kernel','mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
clasif=RandomizedSearchCV(svm.SVC(gamma='auto'),{'kernel':['rbf','linear'],'C':[1,10,20]},cv=5,return_train_score=False,n_iter=2)
clasif.fit(df.drop(columns=['target']),df.target)
results1=pd.DataFrame(clasif.cv_results_)[['param_C','param_kernel','mean_test_score']]
results1
```
