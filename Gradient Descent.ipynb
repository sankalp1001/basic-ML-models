Finding the best curve that fits:

Considering linear regression, we first start with a random line and calculate the error.

y=mx+b

Error calculation is done through many methods.some of them are mean sqaured error, mean absolute error etc. 

The partial derivatives of the error function with respect to all the parameters(coefficient and intercept) is taken. this multiplied with something called the learning rate(decides at what rate the parameters have to change in each case) and finally this value is substracted from the parameter to get the new value.

m=m-learning_rate*de/dm

b=b-learning_rate*de/db

this method is repeated untill the error value doesnt further decrease after updating the weights.

this method is called the Gradient descent   


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y):
    
    learning_rate=0.0001
    
    iterations=10000
    
    n=len(x)
    
    m_curr=b_curr=0
    
    
    for i in range(iterations):
        
        y_pred = m_curr*x + b_curr
        
        cost=(1/n)*sum([val**2 for val in (y-y_pred)])
       
        md=-(2/n) * sum(x*(y-y_pred))
        
        bd=-(2/n) * sum(y-y_pred)
        
        m_curr = m_curr  - learning_rate*md
        
        b_curr = b_curr  - learning_rate*bd
        
        print('m {},b {},cost {},iteration {}'.format(m_curr,b_curr,cost,i))
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
gradient_descent(x,y)
```

first start off with some low iterations and check what learning rate is causing the cost value to decrease. once you find the best learning rate, you can increase the number of iterations for the best result.
